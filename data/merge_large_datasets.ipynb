```python
# importing modules and packages
import pandas as pd
import numpy as np, os

# import libraries for plotting
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use('ggplot')
pd.set_option('display.max_columns', 200)

# ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# prompt: read csv without headers
df = pd.read_csv(resolve_path_gdrive('data/original-datasets/cat_desc.csv', localPathPrefix="../"), header=None)
df_ddg = pd.read_csv(resolve_path_gdrive('data/original-datasets/Data.csv', localPathPrefix="../"), header=None)
# only read catalyst and ddg columns and ignore the other cols
df_ddg = df_ddg[[0, 4]]
df = df.set_index(list(df.columns[[0]]))
df_ddg = df_ddg.set_index(list(df_ddg.columns[[0]]))
df_ddg.columns = ['ddG']
```

```python
# prompt: intersect the 2 df based on the first column
df_intersect = pd.merge(df, df_ddg, how='inner', left_index=True, right_index=True)
#df_intersect = df_intersect.set_index(list(df_intersect.columns[[0]]))
```

```python
df_intersect.to_csv('merged_large_catalyst.csv', index_label = "Catalyst")
```
